**Cleaning Duplicate Data**

There are also duplicate data in the Google Play Store data set according to the [discussions](https://www.kaggle.com/lava18/google-play-store-apps/discussion/76316). 

```python
individual_data = []
duplicate_data = []
for row in android_data:
    app_name = row[0]
    if app_name in individual_data:
        duplicate_data.append(app_name)
    elif app_name not in individual_data:
        individual_data.append(app_name)
```

```python
print("# of duplicate apps:", len(duplicate_data))
print('\n')
print('Examples:',duplicate_data[:11])
```

```python
print(android_data_header)
print('\n')
for row in android_data:
    name_app = row[0]
    if name_app == 'Instagram':
        print(row)
```

As seen above, there are multiple duplicates of the same app. However, we won't be deleting duplicates randomly. If we observe the 'Reviews' category, the highest entry is '66577446'. We will choose this entry since it represents the latest data on the app.

```python
expected_data_length = len(android_data) - len(duplicate_data)
print('Expected Google Play Data Set Length: ', expected_data_length)
```

```python
reviews_max = {} #each app with max reviews 

for row in android_data:
    name = row[0]
    n_reviews = float(row[3])
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name] = n_reviews
    elif name not in reviews_max:
        reviews_max[name] = n_reviews
```

```python
print(expected_data_length == len(reviews_max)) #proceed if TRUE, otherwise review code & dictionary
```

We will now use "reviews_max" dictionary to clean the android data set of the duplicate apps and create a new data set.

```python
android_clean = [] #new data set without duplicates
already_added = [] #this data set helps to avoid duplicate apps with same number of reviews

for row in android_data:
    name = row[0]
    n_reviews = float(row[3])
    if (n_reviews == reviews_max[name]) and (name not in already_added):
        android_clean.append(row)
        already_added.append(name)
```

```python
print(len(android_clean))
print('\n')
print('Expected data length equal to Cleaned Data Set Length: ', expected_data_length == len(android_clean))
```

The data set "android_clean" contains the new data set free of duplicate apps. This is verified as the length of the "android_clean" is equal to the expected length. More obervation on the new data set below.

```python
print(android_data_header)
print('\n')
print(explore_data(android_clean,0,3,True))
```

**Cleaning non English Data**

There are non English app data in the "android_clean" data set. The company we are analyzing the data for is focusing on an English speaking audience. Therefore, we do not need these non English app data and we will remove them. Few examples below:

```python

print(apple_data[441][2]) #app name in 3rd column for apple store data set
print(apple_data[6580][2])
print('\n')
print(android_clean[4412][0]) #app name in 1st column for google play store data set
print(android_clean[7940][0])
```

All these characters that are specific to English texts are encoded using the ASCII standard. Each ASCII character has a corresponding number between 0 and 127 associated with it, and we can take advantage of that to build a function that checks an app name and tells us whether it contains non-ASCII characters.

We will use this to create a function to sort out the non English characters and hence the non English apps. 

```python
def char_def(string):
    for character in string:
        if ord(character) > 127:
            return False
        
    return True

#testing the char_def function
print(char_def('Instagram'))
print(char_def('爱奇艺PPS -《欢乐颂2》电视剧热播'))
print(char_def('Docs To Go™ Free Office Suite'))
print(char_def('Instachat 😜'))
```

```python
print(ord('™'))
print(ord('😜'))
```

Currently, the function to check for non English apps will remove some English apps with use other symbols in thier names such as ™, — , –, etc. These fall outside the ASCII range. 
To avoid loss of useful data, we will update the function to only remove an app if the name has more than 3 non English characters.

```python
def char_def(string):
    non_eng = 0
    for character in string:
        if ord(character) > 127:
            non_eng += 1
        
    if non_eng > 3:
            print(non_eng)
            return False
    
    else:
            print(non_eng)
            return True

print(char_def('Docs To Go™ Free Office Suite'))
print(char_def('Instachat 😜'))
print(char_def('爱奇艺PPS -《欢乐颂2》电视剧热播'))
```

The function might still remove English apps with over 3 non_ASCII characters but the loss of data will be minimal.

Below we use the updated "char_def()" function to clean the data sets of non English apps.

```python
apple_english = []

for row in apple_data:
    name = row[2]
    if char_def(name):
        apple_english.append(row)

        
```

Because our end goal is to add the app on both Google Play and the App Store, we need to find app profiles that are successful on both markets. For instance, a profile that works well for both markets might be a productivity app that makes use of gamification.

---
We will be creating functions to analyze the "prime_genre" column Apple data set. For the Android data set, we will be analyzing the "Category" & "Genres" columns. 

```python
def freq_table(dataset,index):
    table = {}
    total = 0
    for row in dataset:
        total +=1
        column = row[index]
        if column in table:
            table[column] += 1
        else:
            table[column] = 1
            
    percentages = {}
    for key in table:
        percent = (table[key]/total)*100
        percentages[key] = percent      
        
    return percentages
```

```python
print(freq_table(apple_free,-5)) #index:-5 is "prime_genre" column in Apple data set.
```

For readability we will create the function "display_function".

```python
def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
```

```python
display_table(apple_free,-5) #index:-5 is "prime_genre" column in Apple data set.
```

## Most Popular Apps by Genre

**Apple Store Popular Genres**

One way to find out what genres are the most popular (have the most users) is to calculate the average number of installs for each app genre. For the Google Play data set, we can find this information in the "Installs" column, but this information is missing for the App Store data set. As a workaround, we'll take the total number of user ratings as a proxy, which we can find in the "rating_count_tot" app.

We created a function to observe the average number of user ratings per app genre on the Apple Store.

```python
apple_genre = freq_table(apple_free,-5)

for genre_key in apple_genre:
    total = 0
    len_genre = 0
    for row in apple_free:
        genre_name = row[-5]
        if genre_name == genre_key:
            rating_tot = float(row[6])
            total += rating_tot
            len_genre += 1
    average = total / len_genre
    print(genre_key,":",average)
```

The categories with the most ratings, and therefore downloads/usages are Navigation, Reference and Social Networking. Lets investigate further to see the distribution of ratings within the categories/genres.

```python
def genre_by_name(string):
    for row in apple_free:
        name = row[2]
        ratings = row[6]
        genre = row[-5]
        if genre == string:
            print(name,":",ratings) 
        
```

**Google Play Store Popular Genres**

We have data about the number of installs for the Google Play market, so we should be able to get a clearer picture about genre popularity. However, the install numbers don't seem precise enough — we can see that most values are open-ended (100+, 1,000+, 5,000+, etc.)

```python
display_table(android_free,5)
```

One problem with this data is that is not precise. For instance, we don't know whether an app with 100,000+ installs has 100,000 installs, 200,000, or 350,000. However, we don't need very precise data for our purposes — we only want to get an idea which app genres attract the most users, and we don't need perfect precision with respect to the number of users.

We're going to leave the numbers as they are, which means that we'll consider that an app with 100,000+ installs has 100,000 installs, and an app with 1,000,000+ installs has 1,000,000 installs, and so on.

To perform computations, however, we'll need to convert each install number to float — this means that we need to remove the commas and the plus characters, otherwise the conversion will fail and raise an error. We'll do this directly in the loop below, where we also compute the average number of installs for each genre (category).

```python
category_dict = freq_table(android_free,1)

for category in category_dict:
    total = 0
    len_category = 0
    for row in android_free:
        category_app = row[1]
        if category_app == category:
            installs = row[5]
            installs = installs.replace('+','')
            installs = installs.replace(',','')
            installs = float(installs)
            total += installs
            len_category += 1
    average = total / len_category
    print(category,":",average)
        
```

```python
def dom_category_apps(string):
    for row in android_free:
        name = row[0]
        category = row[1]
        installs = row[5]
        if category == string and (installs == '1,000,000,000+' or installs == '500,000,000+' or installs == '100,000,000+'):
            print(name, ':', installs)
```
